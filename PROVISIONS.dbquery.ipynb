```sql
-- SQL for PF provisions, these are only available in databricks and therefore we need to use this query to build. I can add this output as a staging table in DW_REV_OPS easily. 

with provi as (
  select
    ps.servicenow_cmdb_ci_password_safe_cloud_resource_group_sys_id as provision_sys_id,
    ps.servicenow_cmdb_ci_password_safe_cloud_resource_group_pathfinder_enabled as sn_pathfinder_enabled,
    ps.servicenow_cmdb_ci_password_safe_cloud_resource_group_used_for as sn_used_for,
    ps.servicenow_cmdb_ci_password_safe_cloud_resource_group_install_status as sn_install_status,
    ps.servicenow_cmdb_ci_password_safe_cloud_resource_group_account_value as sn_acct_id,
    ps.servicenow_cmdb_ci_password_safe_cloud_resource_group_model_category_value as sn_base_id,
    ps.servicenow_cmdb_ci_password_safe_cloud_resource_group_sys_class_name as sn_sys_class_name,
    ps.servicenow_cmdb_ci_password_safe_cloud_resource_group_sys_created_on_date as sys_created_date,
    ps.servicenow_cmdb_ci_password_safe_cloud_resource_group_sys_updated_on_date as sys_updated_date,
    ps.servicenow_cmdb_ci_password_safe_cloud_resource_group_start_date as start_date,
    split_part(ps.servicenow_cmdb_ci_password_safe_cloud_resource_group_fqdn, '.', 1) as u_resource_key
  from prod_dm.alteryx.alteryx_servicenow_u_cmdb_ci_password_safe_cloud_resource_group ps
  where ps.servicenow_cmdb_ci_password_safe_cloud_resource_group_sys_id is not null
  union all
  select
    pm.servicenow_cmdb_ci_privilege_management_cloud_sys_id as provision_sys_id,
    pm.servicenow_cmdb_ci_privilege_management_cloud_pathfinder_enabled as sn_pathfinder_enabled,
    pm.servicenow_cmdb_ci_privilege_management_cloud_used_for as sn_used_for,
    pm.servicenow_cmdb_ci_privilege_management_cloud_install_status as sn_install_status,
    pm.servicenow_cmdb_ci_privilege_management_cloud_account_value as sn_acct_id,
    pm.servicenow_cmdb_ci_privilege_management_cloud_model_id_value as sn_base_id,
    pm.servicenow_cmdb_ci_privilege_management_cloud_sys_class_name as sn_sys_class_name,
    pm.servicenow_cmdb_ci_privilege_management_cloud_sys_created_on_date as sys_created_date,
    pm.servicenow_cmdb_ci_privilege_management_cloud_sys_updated_on_date as sys_updated_date,
    pm.servicenow_cmdb_ci_privilege_management_cloud_start_date as start_date,
    split_part(pm.servicenow_cmdb_ci_privilege_management_cloud_fqdn, '.', 1) as u_resource_key
  from prod_dm.alteryx.alteryx_servicenow_u_cmdb_ci_privilege_management_cloud pm
  where pm.servicenow_cmdb_ci_privilege_management_cloud_sys_id is not null
  union all
  select
    ra.servicenow_cmdb_ci_privileged_remote_access_cloud_instance_sys_id as provision_sys_id,
    ra.servicenow_cmdb_ci_privileged_remote_access_cloud_instance_pathfinder_enabled as sn_pathfinder_enabled,
    ra.servicenow_cmdb_ci_privileged_remote_access_cloud_instance_used_for as sn_used_for,
    ra.servicenow_cmdb_ci_privileged_remote_access_cloud_instance_install_status as sn_install_status,
    ra.servicenow_cmdb_ci_privileged_remote_access_cloud_instance_account_value as sn_acct_id,
    ra.servicenow_cmdb_ci_privileged_remote_access_cloud_instance_model_id_value as sn_base_id,
    ra.servicenow_cmdb_ci_privileged_remote_access_cloud_instance_sys_class_name as sn_sys_class_name,
    ra.servicenow_cmdb_ci_privileged_remote_access_cloud_instance_sys_created_on_date as sys_created_date,
    ra.servicenow_cmdb_ci_privileged_remote_access_cloud_instance_sys_updated_on_date as sys_updated_date,
    ra.servicenow_cmdb_ci_privileged_remote_access_cloud_instance_start_date as start_date,
    split_part(ra.servicenow_cmdb_ci_privileged_remote_access_cloud_instance_fqdn, '.', 1) as u_resource_key
  from prod_dm.alteryx.alteryx_servicenow_u_cmdb_ci_privileged_remote_access_cloud_instance ra
  where ra.servicenow_cmdb_ci_privileged_remote_access_cloud_instance_sys_id is not null
  union all
  select
    rs.servicenow_cmdb_ci_remote_support_cloud_instance_sys_id as provision_sys_id,
    rs.servicenow_cmdb_ci_remote_support_cloud_instance_pathfinder_enabled as sn_pathfinder_enabled,
    rs.servicenow_cmdb_ci_remote_support_cloud_instance_used_for as sn_used_for,
    rs.servicenow_cmdb_ci_remote_support_cloud_instance_install_status as sn_install_status,
    rs.servicenow_cmdb_ci_remote_support_cloud_instance_account_value as sn_acct_id,
    rs.servicenow_cmdb_ci_remote_support_cloud_instance_model_id_value as sn_base_id,
    rs.servicenow_cmdb_ci_remote_support_cloud_instance_sys_class_name as sn_sys_class_name,
    rs.servicenow_cmdb_ci_remote_support_cloud_instance_sys_created_on_date as sys_created_date,
    rs.servicenow_cmdb_ci_remote_support_cloud_instance_sys_updated_on_date as sys_updated_date,
    rs.servicenow_cmdb_ci_remote_support_cloud_instance_start_date as start_date,
    split_part(rs.servicenow_cmdb_ci_remote_support_cloud_instance_fqdn, '.', 1) as u_resource_key
  from prod_dm.alteryx.alteryx_servicenow_u_cmdb_ci_remote_support_cloud_instance rs
  where rs.servicenow_cmdb_ci_remote_support_cloud_instance_sys_id is not null
  union all
  select
    ep.servicenow_entitle_provisions_sys_id as provision_sys_id,
    ep.servicenow_entitle_provisions_u_pathfinder_enabled as sn_pathfinder_enabled,
    ep.servicenow_entitle_provisions_u_used_for as sn_used_for,
    ep.servicenow_entitle_provisions_install_status as sn_install_status,
    ep.servicenow_entitle_provisions_u_account_value as sn_acct_id,
    ep.servicenow_entitle_provisions_model_number as sn_base_id,
    ep.servicenow_entitle_provisions_sys_class_name as sn_sys_class_name,
    ep.servicenow_entitle_provisions_sys_created_on as sys_created_date,
    ep.servicenow_entitle_provisions_sys_updated_on as sys_updated_date,
    ep.servicenow_entitle_provisions_start_date as start_date,
    ep.servicenow_entitle_provisions_u_resource_key as u_resource_key
  from prod_dm.alteryx.alteryx_servicenow_u_entitle_provisions ep
  where ep.servicenow_entitle_provisions_sys_id is not null
  union all
  select
    i.servicenow_is_platform_provisions_sys_id as provision_sys_id,
    i.servicenow_is_platform_provisions_pathfinder_enabled as sn_pathfinder_enabled,
    i.servicenow_is_platform_provisions_used_for as sn_used_for,
    i.servicenow_is_platform_provisions_install_status as sn_install_status,
    i.servicenow_is_platform_provisions_account_value as sn_account_id,
    null as sn_base_id,
    i.servicenow_is_platform_provisions_sys_class_name as sn_sys_class_name,
    i.servicenow_is_platform_provisions_sys_created_on_date as sn_sys_created_date,
    i.servicenow_is_platform_provisions_sys_updated_on_date as sn_sys_updated_date,
    i.servicenow_is_platform_provisions_start_date as start_date,
    i.servicenow_is_platform_provisions_resource_key as u_resource_key
  from prod_dm.alteryx.alteryx_servicenow_u_is_platform_provisions i
  where i.servicenow_is_platform_provisions_sys_id is not null
  union all
  select
    pm.servicenow_pmlc_platform_provisions_sys_id as provision_sys_id,
    pm.servicenow_pmlc_platform_provisions_u_pathfinder_enabled as sn_pathfinder_enabled,
    pm.servicenow_pmlc_platform_provisions_used_for as sn_used_for,
    pm.servicenow_pmlc_platform_provisions_install_status as sn_install_status,
    pm.servicenow_pmlc_platform_provisions_u_account_value as sn_acct_id,
    pm.servicenow_pmlc_platform_provisions_model_number as sn_base_id,
    pm.servicenow_pmlc_platform_provisions_sys_class_name as sn_sys_class_name,
    pm.servicenow_pmlc_platform_provisions_sys_created_date as sn_sys_created_date,
    pm.servicenow_pmlc_platform_provisions_sys_updated_date as sn_sys_updated_date,
    pm.servicenow_pmlc_platform_provisions_start_date as start_date,
    pm.servicenow_pmlc_platform_provisions_u_resource_key as u_resource_key
  from prod_dm.alteryx.alteryx_servicenow_u_pmlc_platform_provisions pm
  where pm.servicenow_pmlc_platform_provisions_sys_id is not null
)
select 
  m.salesforce_account_id as sfdc_acct_id, p.*, case when pf.salesforce_pathfinder_deployment_service_now_cloud_instance_id_c is not null then 'SFDC Prov' else 'No SFDC Prov' end sfdc_pf_provision
from provi p 
left join prod_dm.alteryx.alteryx_master_account m on p.sn_acct_id = m.servicenow_customer_account_sys_id
left join prod_dm.alteryx.alteryx_salesforce_pathfinder_deployment_c pf on pf.salesforce_pathfinder_deployment_service_now_cloud_instance_id_c = p.provision_sys_id;
```
